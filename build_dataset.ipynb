```python
import fiftyone as fo
import fiftyone.zoo as foz
from fiftyone import ViewField as F
import os
```

```python
export_oi = './export_oi'
data_oi = './data/'
```

```python
recyclables = ["Bottle", "Can", "Carton"]
hazardous = ["Battery"]
kitchen = ["Banana", "Apple", "Orange"]
residual = ["Plastic bag", "Tissue paper", "Cup"]
wanted_classes = recyclables + hazardous + kitchen + residual
```

```python
# 下载 Waste Materials 数据集
dataset = foz.load_zoo_dataset(
    "open-images-v6",
    split="train", 
    label_field=["detection"],
    only_matching=True,
    classes=wanted_classes,
    max_samples=100,  # 限制样本数量
    dataset_name="oi-trashes-train-100", 
    dataset_dir=data_oi,  # 指定保存路径
)
```

```python
session = fo.launch_app(dataset, auto=False, browser=False, port=5151, address="localhost")
print("FiftyOne server 已启动: http://localhost:5151")
```
