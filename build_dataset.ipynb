```python
import fiftyone as fo
import fiftyone.zoo as foz
from fiftyone import ViewField as F
import os
```

```python
export_dir = './export_oi'
```

```python
# 获取有效类别
# fo.utils.openimages.get_classes(version='v6')
```

```python
recyclables = ["Bottle", "Book"]
hazardous = ["Mobile phone"]
kitchen = ["Banana", "Apple", "Orange"]
residual = ["Plastic bag", "Toilet paper", "Coffee cup"]
wanted_classes = recyclables + hazardous + kitchen + residual
# Ignoring invalid classes ['Can', 'Carton', 'Battery', 'Tissue paper', 'Cup']
# No segmentations exist for classes ['Lamp', 'Tin can']
```

```python
# foz.delete_zoo_dataset("open-images-v6", split="train")
```

```python
# if fo.dataset_exists("oi-trashes-train-1000"):
#     fo.delete_dataset("oi-trashes-train-1000")
```

```python
train_dataset = foz.load_zoo_dataset(
    "open-images-v6",
    split="train", 
    label_field="detections",
    only_matching=True,
    classes=wanted_classes,
    max_samples=1000,  # 限制样本数量
    dataset_name="oi-trashes-train-1000", 
    # overwrite=True,
)
```

```python
session = fo.launch_app(train_dataset, auto=False, browser=False, port=5151, address="localhost")
print("FiftyOne server 已启动: http://localhost:5151")
```

```python
train_dataset.export(
    export_dir=export_dir,
    dataset_type=fo.types.YOLOv5Dataset,
    label_field="detections_detections",
    classes=wanted_classes,
    split="train"
)
```

```python
valid_dataset = foz.load_zoo_dataset(
    "open-images-v6",
    split="validation", 
    label_field="detections",
    only_matching=True,
    classes=wanted_classes,
    max_samples=1000,  # 限制样本数量
    dataset_name="oi-trashes-valid-1000", 
    # overwrite=True,
)
```

```python
valid_dataset.export(
    export_dir=export_dir,
    dataset_type=fo.types.YOLOv5Dataset,
    label_field="detections_detections",
    classes=wanted_classes,
    split="val"
)
```
